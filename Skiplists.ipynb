```python
import random


class Listnode(): # creating a standard listnode
    def __init__(self,key,level):
        self.key = key
        self.next = [None] * (level + 1) # creating references for level 
        
class skiplist():
    def __init__(self,maxlevel):
        self.max = maxlevel
        self.head = self.createnode(-1,self.max) # head for starting node
        self.level = 0 # current level 
        
    def createnode(self,key,level):
        node = Listnode(key,level)
        return node         
    
    def randomize_level(self):
        leve = 0
        while random.choice([0,1]) == 1 and leve < self.max:
            print(random.choice([0,1]))
            leve += 1
        return leve
            
    def lookup(self,key):
        curr = self.head
        foundlevel = -1
        for i in range(self.level, -1,-1):
            while(curr.next[i] and curr.next[i].key < key):
                print("going right")
                curr = curr.next[i]
            print(f"Going down to level:{i}")

            if curr.next[i] and curr.next[i].key == key:
                foundlevel = i
                break
        if foundlevel != -1:
            print(f"Key found: {key} on the level -> {i}")

        else:
            print("not found")
        
                
        
    def insert(self,key):
        curr = self.head 
        totallist = [None]*(self.max + 1)
        for i in range(self.level,-1,-1):
            while curr.next[i] and curr.next[i].key < key: # stop before key 
                curr = curr.next[i] # keep moving in the linked list
            totallist[i] = curr
                
        curr = curr.next[0] # at the bottom level/ linked list

        if curr == None or curr.key is not key:
            randomlevel = self.randomize_level()
            if randomlevel > self.level:
                for i in range(self.level + 1,randomlevel + 1):
                    totallist[i] = self.head
                self.level = randomlevel # We're at the desired level
            new = self.createnode(key,randomlevel) # made a new node in that level

            for i in range(randomlevel + 1):
                new.next[i] = totallist[i].next[i]
                totallist[i].next[i] = new 
            print(f"inserted key {key}")
            
    def delete(self,key):
        totallist = [None]*(self.max + 1)
        curr = self.head

        for i in range(self.level, -1,-1):
            while(curr.next[i] and curr.next[i].key < key):
                curr = curr.next[i]
            totallist[i] = curr

        curr = curr.next[0]

        if curr is not None and curr.key == key:
            for i in range(self.level + 1):

                if totallist[i].next[i] is not curr:
                    break
                totallist[i].next[i] = curr.next[i]
            
            while(self.level > 0 and self.head.next[self.level] == None):
                self.level -= 1
                print("Deleted {}".format(key))

    def printlist(self):
        
        start = self.head

        for i in range(self.max+1):
            print(f"The level -> {i}:", end= " ")
            node = start.next[i]
            while (node is not None):
                print(node.key,end= " ")
                node = node.next[i] 
            print("")

```
