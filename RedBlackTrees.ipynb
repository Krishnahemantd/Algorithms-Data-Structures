```python
class Treenode():
    def __init__(self,key):
        self.key = key
        self.p = None
        self.left = None
        self.right = None
        self.color = 1 # Red value

class RBtree():

    def __init__(self):
            self.Nil = Treenode(0) # Black value
            self.Nil.color = 0
            self.Nil.left = None
            self.Nil.right = None
            self.root = self.Nil

    def Search(self,k):
        
        node = self.root
     
        while node != self.Nil and k != node.key:
            if k < node.key:
               node = node.left 
            else:
               node = node.right
        
        return node 
    

    def minimum(self,k):
        node = self.Search(k)
        while node.left != self.Nil:
            node = node.left 
        return node.key

    
    def maximum(self,k):
        node = self.Search(k)
        while node.right != self.Nil:
            node = node.right
        return node.key 

    def Successor(self,k):
        node = self.Search(k)
        if node == self.Nil:
            return None
        if node.right != self.Nil:
            return self.minimum(node.right.key)
        else:
            y = node.p
            while y != self.Nil and node == y.right:
                node = y
                y = y.p
            return y 

    def Predecessor(self, k):
        node = self.Search(k)
        if node == self.Nil:
            return None 
        if node.left != self.Nil:
            return self.maximum(node.left.key)
        else:
            y = node.p
            while y != self.Nil and node == y.left:
                node = y
                y = y.p
            return y
        
    def Searchmain(self, k):
        nodes = self.Search(k)
        node_height = self.height(nodes)

        if nodes:
            hh = self.height(self.root) - node_height
            print(f"Node found: {nodes.key} at level: {hh}")
        else:
            print("Node not found")

        
    def height(self,node):
        if node == self.Nil:
            return 0

        h = 0
        currlevel = [node]

        while currlevel:
            
            nextl = []

            for node in currlevel:
                if node.left != self.Nil:
                    nextl.append(node.left)
                if node.right != self.Nil:
                    nextl.append(node.right)

            currlevel = nextl
            h += 1

        #print(f"The height is :{h}")
        return h
    
    def Treeheight(self):

        print(f"The height of the tree is :{self.height(self.root)}")

    def sort(self):
        sorted_list = []
        self.inorder(self.root, sorted_list)
        return sorted_list

    def inorder(self, node, sorted_list):
        if node is not self.Nil:
            self.inorder(node.left, sorted_list)
            sorted_list.append(node.key)  
            self.inorder(node.right, sorted_list)

    def LeftRotate(self,x):
        y = x.right
        x.right = y.left
        if y.left != self.Nil:
                y.left.p = x
        y.p = x.p

        if x.p == None:
            self.root = y
        
        elif x == x.p.left:
             x.p.left = y

        else:
            x.p.right = y
        
        y.left = x
        x.p = y

    def RightRotate(self,x):
        y = x.left
        x.left = y.right
        if y.right != self.Nil:
                y.right.p = x
        y.p = x.p

        if x.p == None:
            self.root = y
        
        elif x == x.p.right:
             x.p.right = y

        else:
            x.p.left = y
        
        y.right = x
        x.p = y
    
    def Insert(self,z):
        node = Treenode(z)
        node.p = None
        node.key = z
        node.left = self.Nil
        node.right = self.Nil
        node.color = 1
        
        
        x = self.root
        y = None
        while x != self.Nil:
            y = x
            if node.key < x.key:
                x = x.left
            else: 
                x = x.right
        node.p = y 
    
        if y == None:
            self.root = node
        elif node.key < y.key:
            y.left = node
        else: 
            y.right = node

        if node.p == None:
            node.color = 0
            return 
        if node.p.p == None:
            return 
    
        self.Insert_Fixup(node)



    def Insert_Fixup(self,z):
        while z.p.color == 1:
            if z.p == z.p.p.left: 
                y = z.p.p.right 
                if y.color == 1:
                     z.p.color = 0
                     y.color = 0
                     z.p.p.color = 1 
                     z = z.p.p

                else:
                    if z == z.p.right:
                        z = z.p
                        self.LeftRotate(z)
                    z.p.color = 0
                    z.p.p.color = 1
                    self.RightRotate(z.p.p)
            else:

                y = z.p.p.left 
                if y.color == 1:
                     z.p.color = 0
                     y.color = 0
                     z.p.p.color = 1
                     z = z.p.p

                else:
                    if z == z.p.left:
                        z = z.p
                        self.RightRotate(z)
                    z.p.color = 0
                    z.p.p.color = 1
                    self.LeftRotate(z.p.p)
            if z == self.root:
                break
        
        self.root.color = 0


    def Transplant(self,u,v):
        if u.p == None:
            self.root = v
        elif u == u.p.left:
            u.p.left = v
        else:
            u.p.right = v
        v.p = u.p

    def Delete(self,key):
        self.DeleteHelp(self.root,key)

    def DeleteHelp(self,node,val):
        z = self.Nil
        while node is not self.Nil:
            if node.key == val:
                z = node 

            if node.key <= val:  # searching for the node as assigning it to z
                node = node.right
            else:
                node = node.left 
            
        if z == self.Nil: # just checking if z is still Nil after the while loop
            print("cant find")
            return 
        
        y = z
        y_original_color = y.color
        if z.left == self.Nil:
            x = z.right 
            self.Transplant(z,z.right)
        elif z.right == self.Nil:
            x = z.left
            self.Transplant(z,z.left)
        else:
            y = self.minimum(z.right)
            y_original_color = y.color 
            x = y.right 
            if y != z.right:
                self.Transplant(y.right)
                y.right = z.right 
                y.right.p = y
            else:
                x.p = y

            self.Transplant(z,y)
            y.left = z.left 
            y.left.p = y
            y.color = z.color 

        print(f"Succesfully deleted {val}")
        if y_original_color == 0:

            self.DeleteFixup(x)
                
    def DeleteFixup(self,x):

        while x != self.root and x.color == 0:
            if x == x.p.left:
                w = x.p.right 

                if w.color == 1:
                    w.color = 0
                    x.p.color = 1
                    self.LeftRotate(x.p)
                    w = x.p.right 
                if w.left.color == 0 and w.right.color == 0:
                    w.color = 1
                    x = x.p
                else:

                    if w.right.color == 0:
                        w.left.color = 0
                        w.color = 1
                        self.RightRotate(w)
                        w = x.p.right
                    w.color = x.p.color 
                    x.p.color = 0
                    w.right.color = 0
                    self.LeftRotate(x.p)
                    x = self.root 

            else:
                w = x.p.left

                if w.color == 1:
                    w.color = 0
                    x.p.color = 1
                    self.RightRotate(x.p)
                    w = x.p.left
                if w.right.color == 0 and w.left.color == 0:
                    w.color = 1
                    x = x.p
                else:

                    if w.left.color == 0:
                        w.right.color = 0
                        w.color = 1
                        self.LeftRotate(w)
                        w = x.p.left
                    w.color = x.p.color 
                    x.p.color = 0
                    w.left.color = 0
                    self.RightRotate(x.p)
                    x = self.root 
        x.color = 0

    def printtree(self):
        if self.root != self.Nil:
            self.printhelper(self.root, 0, "")
        else:
            print("Empty tree")

    def printhelper(self, node, level, direction):
        if node != self.Nil:
            self.printhelper(node.right, level + 1, "R")
            print("     " * level + "+--" + direction + f"({node.key}, {'Red' if node.color == 1 else 'Black'}) --Level: {level}")
            self.printhelper(node.left, level + 1, "L")

                    
```
